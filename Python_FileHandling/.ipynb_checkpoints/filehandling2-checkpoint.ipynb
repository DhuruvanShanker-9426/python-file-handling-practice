```python
with open("sports.txt","x") as sp:
    sp.write("cricket\n")
    sp.write("chess\n")
    sp.write("baseball\n")
    sp.write("volleyball\n")
```

```python
with open("sports.txt","r") as read_sports_file:
    res=read_sports_file.read()
    print(res)
    print(len(res))
```

```python
with open("sports.txt","r") as rlsp:
    ls=rlsp.readlines()
    print(ls)
    print(len(ls))
    print("lines------------------")
    for i in ls:
        print(len(i))
```

# Write Mode

```python
with open("sports.txt","w") as wr_sp:
    wr_sp.write("kabadi")
    wr_sp.write("shogi")  #shogi , kabadishogi
```

```python
with open("sports.txt","w") as wr_sp:
    wr_sp.write("tennis")
    wr_sp.write("basketball")
```
